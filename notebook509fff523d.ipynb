```python
from ultralytics import YOLO

# Load your trained model
model = YOLO("runs/detect/train/weights/best.pt")

# Run tracking
results = model.track(
    source="/kaggle/input/datasets/atishayjain097/tt-dataset/videoplayback (2).mp4",
    conf=0.3,
    tracker="bytetrack.yaml",
    save=True
)
```

```python
import math

ball_positions = []

frame_idx = 0  # manually track frame number

for r in results:
    if r.boxes.id is None:
        frame_idx += 1
        continue

    boxes = r.boxes.xyxy.cpu().numpy()
    ids = r.boxes.id.cpu().numpy()

    for box, track_id in zip(boxes, ids):
        x1, y1, x2, y2 = box
        
        cx = (x1 + x2) / 2
        cy = (y1 + y2) / 2
        
        # store (frame_number, ball_id, x, y)
        ball_positions.append((frame_idx, int(track_id), cx, cy))

    frame_idx += 1
```

```python
from collections import defaultdict

# ball_positions = [(frame, track_id, cx, cy), ...]

tracks = defaultdict(list)

for frame, track_id, x, y in ball_positions:
    tracks[track_id].append((frame, x, y))
```

```python
# =============================
# SCALE & VIDEO PARAMETERS
# =============================

# Frames per second of the video
FPS = 30   # change to 60 if your video is 60fps

# Real-world table length (meters)
TABLE_LENGTH_METERS = 2.74  # official table tennis table length

# Approximate table length in pixels (measure once visually)
TABLE_LENGTH_PIXELS = 600   # adjust based on your video

# Pixel ‚Üí meter conversion
meters_per_pixel = TABLE_LENGTH_METERS / TABLE_LENGTH_PIXELS

print("Meters per pixel:", meters_per_pixel)
```

```python
ball_speeds = []

for track_id, points in tracks.items():
    points = sorted(points, key=lambda x: x[0])  # sort by frame

    for i in range(1, len(points)):
        f1, x1, y1 = points[i - 1]
        f2, x2, y2 = points[i]

        pixel_dist = ((x2 - x1)**2 + (y2 - y1)**2) ** 0.5
        meter_dist = pixel_dist * meters_per_pixel

        time_sec = (f2 - f1) / FPS
        if time_sec == 0:
            continue

        speed_mps = meter_dist / time_sec
        speed_kmph = speed_mps * 3.6

        ball_speeds.append((f2, speed_kmph))
```

```python
import numpy as np

frames = [f for f, s in ball_speeds]
speeds = [s for f, s in ball_speeds]

window = 5
smooth_speeds = np.convolve(speeds, np.ones(window)/window, mode="valid")
smooth_frames = frames[:len(smooth_speeds)]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(smooth_frames, smooth_speeds)
plt.xlabel("Frame Number")
plt.ylabel("Ball Speed (km/h)")
plt.title("Table Tennis Ball Speed Over Time")
plt.grid(True)
plt.show()
```
