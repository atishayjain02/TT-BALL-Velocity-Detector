```python
import os
import shutil

# Source frames folder
src = "/kaggle/working/frames"

# Destination sampled frames folder
dst = "/kaggle/working/frames_sampled"

# Create destination folder
os.makedirs(dst, exist_ok=True)

frames = sorted(os.listdir(src))

for i, f in enumerate(frames):
    if i % 3 == 0:  # keep every 3rd frame
        shutil.copy(
            os.path.join(src, f),
            os.path.join(dst, f)
        )

print("Total original frames:", len(frames))
print("Sampled frames:", len(os.listdir(dst)))
```

```python
import shutil

shutil.make_archive(
    "/kaggle/working/frames_sampled",
    'zip',
    "/kaggle/working/frames_sampled"
)

print("ZIP created: frames_sampled.zip")
```
